```python
import pandas as pd
```

```python
xlsx_file = '/Users/educa/Documents/STUDY/GOPAS/subory/DATA/ukraine-support.xlsx'
```

Funkcia pandas.read_excel() vyžaduje nainštalovanú knižnicu openpyxl (alebo iný podporovaný engine), aby mohla načítať excelovský súbor
Pri zavolaní funkcie read_excel() pandas najprv skontroluje, či je nainštalovaný engine a použije ho na načítanie zo súboru.
Ak nie je nainštalovaný, vyvolá chybu ImportError

```python
df = pd.read_excel(xlsx_file)
df.head(5)
```

dáta sú však v druhom hárku, automaticky načítal prvý. Ak chceme iný, tak to treba zadať do parametrov
